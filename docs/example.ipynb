# Example usage

To use `pycounts_rc` in a project:

```python
import pycounts_rc

print(pycounts_rc.__version__)
```

Here we will demonstrate how to use the words ina  text file and plot the top 5 results. 


## Imports

```python
from pycounts_rc.pycounts_rc import count_words
from pycounts_rc.plotting import plot_words
```

## Create a text file 

We'll first create a text file to work with using a famous quote from Einstein: 

```python
quote = """Insanity is doing the same thing over and over again expecting different results.""" 
with open("einstein.txt", "w") as file: 
    file.write(quote) 
```

## Count words 

We can count the words in our file using the 'count_words()` fuinction. Note that this function removes punctuation and makes all words lowercase before counting. 

```python
counts = count_words("einstein.txt") 
print(counts) 
```

## Plot words

We can now plot the result using the `plot_words()` function: 

```python
fig = plot_words(counts, n=5) 
```
